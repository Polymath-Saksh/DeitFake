```python
IMAGE_ROOT = "/content/OpenForensics_TestDev/images"

def resolve_image_path(json_path):
    return os.path.join(IMAGE_ROOT, os.path.basename(json_path))
```

```python
import os
def find_image(filename):
    for root in IMAGE_ROOT:
        path = os.path.join(root, filename)
        if os.path.exists(path):
            return path
    raise FileNotFoundError(filename)
```

```python
import json

with open("/content/Test-Dev_poly.json") as f:
    test_dev = json.load(f)

# Look at first 5 filenames
for img in test_dev["images"][:5]:
    print(img["file_name"])
```

```python
image_id_to_file = {
    img["id"]: img["file_name"]
    for img in test_dev["images"]
}
```

```python
from collections import defaultdict

image_id_to_anns = defaultdict(list)
for ann in test_dev["annotations"]:
    image_id_to_anns[ann["image_id"]].append(ann)
```

```python
def crop_face(image, bbox):
    x, y, w, h = bbox
    x1 = max(0, x)
    y1 = max(0, y)
    x2 = min(image.width, x + w)
    y2 = min(image.height, y + h)
    return image.crop((x1, y1, x2, y2))
```

```python
from transformers import AutoImageProcessor, AutoModelForImageClassification
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

processor = AutoImageProcessor.from_pretrained("sakshamkr1/deitfake-v2", use_fast=True)
model = AutoModelForImageClassification.from_pretrained(
    "sakshamkr1/deitfake-v2"
).to(device)
model.eval()
```

## Face-Level Prediction

```python
import torch.nn.functional as F

def predict_face(face_img):
    inputs = processor(images=face_img, return_tensors="pt").to(device)
    with torch.no_grad():
        logits = model(**inputs).logits
        probs = torch.softmax(logits, dim=-1)
    return probs[0, model.config.label2id["Fake"]].item()
```

```python
def aggregate(scores):
    return max(scores)
```

```python
from PIL import Image
from tqdm import tqdm

y_true = []
y_score = []

for img_id, anns in tqdm(image_id_to_anns.items()):
    rel_path = image_id_to_file[img_id]
    img_path = resolve_image_path(rel_path)

    image = Image.open(img_path).convert("RGB")

    face_scores = []
    for ann in anns:
        face_crop = crop_face(image, ann["bbox"])
        score = predict_face(face_crop)
        face_scores.append(score)

    if not face_scores:
        continue

    img_score = aggregate(face_scores)
    y_score.append(img_score)

    # Image-level GT: fake if ANY face is fake
    img_label = int(any(a["category_id"] == 1 for a in anns))
    y_true.append(img_label)
```

```python
from collections import Counter
print("Label distribution:", Counter(y_true))
```

```python
import numpy as np

y_pred = (np.array(y_score) >= 0.5).astype(int)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_true, y_score)
print("OpenForensics Test-Dev AUROC:", auc)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_true, y_pred)
print(f"Accuracy: {acc:.4f}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)
```
